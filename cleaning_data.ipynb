```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
df.info()
```

```python
# To see how many duplicated rows there are
df = df.drop_duplicates() # To remove duplicated rows
```

```python
#convertir tipos de fecha
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['MONTH_ID'] = pd.to_datetime(df['MONTH_ID'], format='%Y-%m')
df['YEAR_ID '] = pd.to_datetime(df['YEAR_ID'], format='%Y')

df.head(2)
```

```python
#manejar valores nulos
df = df.fillna(df.mode().iloc[0])#rellenar nulos categoricos con la moda
```

```python
#veriicar nulos en porcentaje
null_percentage = df.isnull().mean() * 100
print(null_percentage)
```

```python
#convertir a minusculas los nombres de las columnas y textos categoricos
df.columns = df.columns.str.lower()
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].str.lower().str.strip()
    
    
df.head(2)
```

```python
#guardar el dataframe limpio
df.to_csv('sales_data_cleaned.csv', index=False)
```
